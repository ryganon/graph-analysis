## Descrição da base utilizada

Para as tarefas de detecção de comunidades descritas nas seções seguintes será utilizada uma base de dados disponibilizada pelo projeto SNAP de Stanford [1]. Os dados foram coletados do site de comércio eletrônico Amazon.com e são referentes à funcionalidade de recomendação de produtos "Clientes que compraram este item também compraram". Sempre que um cliente estiver visualizando a página de um determinado produto, uma seção no fim da página apresenta uma lista de produtos que outros clientes compraram em conjunto com o produto da página atual. É o caso de um sistema de recomendação de produtos baseado em compras já realizadas.

Para a construção da base segue-se o padrão:

* Os produtos são representados como nós no grafo.
* Se um determinado produto $A$ é comprado em conjunto com o produto $B$, então o grafo contém uma ligação não-direcionada entre os nós $A$ e $B$.

Assume-se que os produtos possuem um conjunto de características. Quando um produto é frequentemente comprado em conjunto com outros produtos, então estes produtos possuem algumas (não necessariamente todas) características em comum. Produtos com características em comum podem, então, ser agrupados em comunidades de produtos. Para demonstrar técnicas de detecção de comunidades, serão utilizados os dados da base para a criação de um grafo não-direcionado. A estrutura do grafo servirá como base para os algoritmos de detecção de comunidades propostos.

A seguir são apresentadas as tarefas de leitura da base de dados e criação do grafo com o framework **NetworkX**.

```python
# leitura do arquivo de dados
data = pd.read_table(os.path.join(PATH,'com-amazon.ungraph.txt'), sep='\t',
                     header=None, skiprows=4, names=['FromNodeId','ToNodeId'])

# criando o grafo não-direcionado
graph = nx.Graph([tuple(x) for x in data.values])

# quantidade de triângulos
graph_triangle = nx.triangles(graph)

print 'Nós:', len(graph.nodes())
print 'Ligações:', len(graph.edges())
print 'Quantidade de triângulos', sum(graph_triangle.values())/3
```

# Analisando a Estrutura do Grafo

Todas as tarefas para a detecção de comunidades apresentadas a seguir são baseadas apenas na estrutura do grafo. Dessa forma, serão considerados apenas os nós e suas ligações. A área de Teoria dos Grafos é dedicada à esse tipo de análise e sua literatura é bastante vasta. Detalhes sobre os conceitos básicos podem ser encontrados no livro *"Networks an Introduction, Newman"* [2].

Algumas das propriedades da estrutura do grafo estão ligadas diretamente ao conceito de formação de comunidades: **cliques**, **k-cores**, **componentes**. E estas propriedades estruturais são utilizadas para técnicas de detecção de comunidades tais como **k-cliques**, **k-cliques communities** (que utiliza o método de percolação) e **clusters hierárquicos**.

## Cliques

**Definição:** Em um grafo não-direcionado, um clique é um subconjunto máximo de vértices tal que cada membro do subconjunto é conectado (por ligações) a todos os outros membros do subconjunto. Um subconjunto máximo indica que não é possível adicionar outros nós ao subconjunto e manter o clique preservado.

A propriedade estrutural do clique em um grafo indica que os nós em questão podem ter alguma forma de similaridade e de coesão entre si. Para uma análise subjetiva, a presença de cliques indicam que os nós são relacionados. Os conjuntos de cliques encontrados em um grafo podem ser sobrepostos, ou seja, um nó pode pertencer à mais de um clique ao mesmo tempo. A característica da sobreposição pode ajudar na tarefa de detecção de comunidades. Uma variação do conceito de cliques é o **k-plex** onde a propriedade de que todos os nós devem estar conectados uns aos outros é relaxada (mais referências, ler [2]).

**Definição:** Um clique em um grafo não-direcionado $G = (V, E)$ é um subconjunto $C$ de vértices, $C \subseteq V$, tal que para cada dois vértices em $C$ existe uma ligação. O subgrafo $C$ é completo (preserva o clique). Os cliques em um grafo podem ter tamanhos variados. Um **clique máximo** é um clique de maior tamanho (quantidade de nós) entre os cliques possíveis de um grafo. O **valor de clique** (clique number), $\omega(G)$, de um grafo $G$ é o número de nós presente em um clique máximo de $G$. 

Para obter todos os cliques de um grafo (serão considerados apenas os cliques máximos) o código é o que segue.

```python
# criando lista com todos os cliques do grafo
list_maximal_cliques = list(nx.clique.find_cliques(graph))

clique_number = nx.clique.graph_clique_number(graph)

# contando a quantidade de cliques por número de nós
counter_cliques = Counter([len(x) for x in list_maximal_cliques])
counter_cliques = DataFrame(counter_cliques.items(), columns=['Qtde. nós','Total encontrado'])

print 'Clique number:', clique_number
print 'Quantidade de cliques encontrados:', len(list_maximal_cliques)
counter_cliques
```

Dentre os $334.863$ nós e $925.872$ ligações foram encontrados $470.042$ cliques com tamanhos que variam entre $2$ e $7$ nós. Cliques com apenas $2$ ou $3$ nós podem não representar uma comunidade de forma significativa então consideraremos para análise posterior apenas os cliques com um número maior que $4$ nós.  

Um exemplo de clique com $7$ nós:

```python
cliques_tam_7 = [x for x in list_maximal_cliques if len(x)==7]
draw_subgraph(cliques_tam_7[0], graph, title=u'Clique com 7 nós', size=(10,8), layout='default') 
```

## k-Cores

Um **k-core** é um subconjunto máximo de nós tal que cada nó é conectado a pelo menos $k$ outros nós no subconjunto. É um conceito derivado dos cliques porém difere no fato de que os nós em k-cores não podem ser sobrepostos, ou seja, um nó não pode pertencer a mais de um k-core.

A técnica de deteção dos k-cores é relativamente simples de ser implementada. Trata-se de um processo recursivo que elimina, em cada iteração, os nós folhas que possuem grau menor que $k$. Obviamente, a eficiência do algoritmo depende do tamanho do grafo e da quantidade de iterações necessárias para obter os subconjuntos finais. 

O **valor do core** (core number) é o maior valor de $k$ (quantidade de ligações) encontrado entre os subgrafos criados. O cálculo do k-core em um grafo é demonstrado a seguir.

```python
# calcula o valor do core number para cada nó do grafo
graph_core_number = nx.algorithms.core_number(graph)

# gera um subgrafo com todos os k-cores possíveis
graph_k_core = nx.algorithms.k_core(graph, core_number=graph_core_number)
```

O resultado final é um grafo com vários subgrafos denominados **componentes**. Um componente é definido como um subconjunto de nós tal que para cada nó do subconjunto existe pelo menos um caminho de ligação para os demais nós. Essa propriedade garante que o componente seja um subconjunto máximo, ou seja, não é possível adicionar novos nós no componente e manter a propriedade de ligação entre os nós. Componentes estão relacionados com outra propriedade dos grafos, a conectividade. Um grafo é **conectado** se, para cada nó, existe ao menos um caminho possível ligando o nó e os demais nós do grafo. O caso contrário define que o grafo, então, é **desconectado**.

É fácil demonstrar que o processo de cálculo de k-cores gera como resultado final um grafo desconectado:

```python
print 'Grafo conectado?', nx.is_connected(graph_k_core)
print 'Total de componentes:', nx.algorithms.number_connected_components(graph_k_core)
```

A execução acima mostra que o algoritmo de k-core detectou um número menor de componentes, $55$, do que o algoritmo de cálculo de cliques. A diferença, para este caso, é que os componentes possuem uma variação de tamanho maior que cliques encontrados. É possível, então, também considerar cada componente como uma comunidade. 

Verificando os componentes gerados, temos:

```python
# listando os componentes do grafo
list_components = list(nx.algorithms.connected_components(graph_k_core))

temp =  Series([len(x) for x in list_components])
print 'Componente com menor número de nós:', temp.min()
print 'Componente com maior número de nós:', temp.max()
```

```python
# exibindo os cinco primeiros componentes do grafo
temp = graph_k_core.subgraph(nbunch= [x for y in list_components[:5] for x in y])
draw_subgraph(temp, graph, size=(20,10), layout='default')
```

# Técnicas para Detecção de Comunidades

Propriedades da estrutura do grafo como as descritas anteriormente podem ser utilizadas para a tarefa de detecção de comunidades. Porém, são conceitos simples e baseados apenas em algumas das propriedades estruturais do grafo.

Conceitualmente, em uma comunidade de nós espera-se a seguinte carecterística: os nós do conjunto devem ter um grande número de conexões entre si e poucas conexões com outras partes do grafo, ou seja, a quantidade de ligações internas deve ser maior que a quantidade de ligações externas. Detectar comunidades é uma tarefa similar à tarefa de **particionamento do grafo** porém diferem na necessidade de determinar a quantidade de partições desejadas. Na detecção de comunidades o número de grupos que devem ser detectados não é especificado, espera-se que as comunidades encontradas sejam formada de forma natural. 

É possível ampliar o conceito de particionamento de grafos e usar como base para novas técnicas aplicadas para detecção de comunidades. Como exemplo, uma forma eficiente de detecção de comunidades pode ser baseada na métrica que verifica o **nível de modularidade** entre as comunidades encontradas.

## Comunidades com k-Cliques

A propriedade dos cliques em um grafo pode ser ampliada e utilizada para a detecção de comunidades. Sabe-se que em um clique todos os nós possuem ligações os demais nós do clique. Porém, quanto maior o grafo maior a chance de encontrar um grande número de cliques. Nas execuções realizadas, a quantidade de nós nos cliques foram variadas os tamanhos, pequenos. Os cliques de maior tamanho no grafo possuem apenas $7$ nós. 

Uma das propriedades do clique pode ser explorada e utilizada para detectar comunidades maiores: a propriedade de que um nó pode pertencer a mais de um clique no grafo. O **método de detecção de comunidades baseado em percolação de cliques** [3] foi desenvolvido com esse propósito. De forma simples, o objetivo é verificar os cliques que possuem uma determinada quantidade de nós que podem ser sobrepostos: cliques que possuem uma quantidade de nós em comum podem ser unidos formando um único conjunto, e consequentemente, maior. O método é baseado no conceito de **k-cliques**.

**Definição:** Um k-clique é um subconjunto máximo de nós tal que para cada nó a distância para os demais nós é de no máximo $k$. A distância entre dois nós é medida pela quantidade de ligações entre os nós do caminho percorrido.  

O algoritmo para o método funciona em três etapas:

- Identificar todos os k-cliques do grafo.
- Dois k-cliques são adjacentes (vizinhos) se eles compartilham pelo menos $k-1$ nós.
- K-cliques adjacentes podem ser sobrepostos (unidos) e então o novo conjunto é a união dos k-cliques adjacentes. 

A detecção utilizando k-cliques é feita a seguir e usa um valor de $k=5$.

```python
# cliques que possuem pelo menos k-1 nós adjacentes.
kc_comm = list(nx.algorithms.k_clique_communities(graph, 5, cliques = list_maximal_cliques))

print 'Total de comunidades:', len(kc_comm)
```

```python
df_kc_comm = DataFrame([[len(x), list(x)] for x in kc_comm], columns=['Qtde. nós','Nós Comunidade'])
df_kc_comm.sort(columns=['Qtde. nós'], ascending=False, inplace=True)
```

```python
print 'Quantidade de nós na maior comunidade:', df_kc_comm['Qtde. nós'].max()
print 'Quantidade de nós na menor comunidade:', df_kc_comm['Qtde. nós'].min()
```

```python
print 'As 10 maiores comunidades encontradas'
df_kc_comm[:10]
```

```python
# grafo com a maior comunidade encontrada com o método k-clique
draw_subgraph(df_kc_comm.iloc[0][1], 
              graph, 
              title=u'Maior comunidade encontrada com o método de percolação e k-clique', 
              size=(20,25))
```

```python
# grafo com a terceira maior comunidade encontrada com o método k-clique (58 nós)
draw_subgraph(df_kc_comm.iloc[3][1], 
              graph, 
              title=u'Comunidade encontrada com o método de percolação e k-clique (58 nós)', 
              size=(20,25))
```

```python
# Imports e configurações gerais
# Necessário: ipython, pandas, networkx, matplotlib

%matplotlib inline
import pandas as pd
import numpy as np
import networkx as nx
import matplotlib.pyplot as plt
import json
import pickle
import operator
import csv
import os

from networkx.algorithms.approximation import clique
from pandas import DataFrame, Series
from networkx.readwrite import json_graph
from IPython.display import IFrame
from IPython.display import FileLink, FileLinks
from collections import Counter
from random import random

PATH = ''  #caminho para os arquivos com os dados das comunidades
```

```python
# Métodos de auxílio

def draw_graph_from_data(nodes, graph, title='', layout='default', size=(6,5)):
    """Método para criar plots com subgrafos de um dado grafo
    Recria o grafo com os nós especificados. Os nós extras serão
    diferenciados por cor
    
    data = array com valores de nós
    graph = grafo original
    """
    edges = graph.edges(nbunch=nodes)
    g = nx.Graph()
    g.add_edges_from(edges)
    
    fig = plt.figure(figsize=size)
    plt.title(title)
    
    if layout == 'circular':
        pos = nx.circular_layout(g)
    elif layout == 'default':
        pos = nx.spring_layout(g)
    
    # os nós passados para a função ficam com cor destacada
    nx.draw(g, pos, nodelist=nodes, node_color='c', node_size=2500, alpha=0.8)
    
    # separa os nós adicionados com a função add_edges e desenha com outra cor
    new_nodes = list(set(g.nodes()) - set(nodes)) 
    nx.draw(g, pos, nodelist=new_nodes, node_color='w', node_size=2500, alpha=0.8)
    nx.draw_networkx_labels(g, pos)
    
def draw_subgraph(nodes, graph, title='', layout='default', colors=False, size=(6,5)):
    """Método para criar plots com subgrafos de um dado grafo, 
    Recria o grafo APENAS com os nós especificados. 
    
    data = array com valores de nós
    graph = grafo original
    """
    g = graph.subgraph(nbunch=nodes)
    fig = plt.figure(figsize=size)
    plt.title(title)
    
    if layout == 'circular':
        pos = nx.circular_layout(g)
    elif layout == 'default':
        pos = nx.spring_layout(g)
    
    if colors:
        colors = [(random(), random(), random()) for i in range(len(g.nodes()))]   
        nx.draw(g, pos, node_color=colors, node_size=2500, alpha=0.8)
    else:
        nx.draw(g, pos, node_color='w', node_size=2500, alpha=0.8)
    
    nx.draw_networkx_labels(g, pos)
    
def draw_graph(graph, title='', layout='default', colors=False, size=(6,5)):
    """Método para criar plots com grafos 
    
    graph = grafo original
    """
    fig = plt.figure(figsize=size)
    plt.title(title)
    
    if layout == 'circular':
        pos = nx.circular_layout(graph)
    elif layout == 'default':
        pos = nx.spring_layout(graph)
    
    if colors:
        colors = [(random(), random(), random()) for i in range(len(g.nodes()))]   
        nx.draw(graph, pos, node_color=colors, node_size=2500, alpha=0.8)
    else:
        nx.draw(graph, pos, node_color='w', node_size=2500, alpha=0.8)
    
    nx.draw_networkx_labels(graph, pos)
```
